```python
#The only data I really need is TMP, date, and station name to link to the Pricing zone
#I am going to start by converting the TMP column to something useful
weather_data['TMP'] = [i[1:-2] for i in weather_data['TMP']]
```

```python
weather_data['TMP'] = weather_data['TMP'].astype(int)
```

```python
def cel_to_fahr(temp_cel):
    temp_fahr = ((temp_cel / 10) * (9/5)) + 32
    return temp_fahr
weather_data['TMP'] = cel_to_fahr(weather_data['TMP'])
weather_data.head()
```

```python
#Now I want to create a column that matches the weather data station names to the pricing zone names
#weather_data['NAME'] = weather_data['NAME'].astype(str)
def label_zone (name):
    if name == 'CHIHUAHUA CHIH., MX':
        return ('CHIHUAHUA')
    if name == 'GUADALAJARA JAL., MX':
        return ('GUADALAJARA')
    if name == 'MEXICO CENTRAL D.F., MX':
        return ('VDM SUR')
    if name == 'MONTERREY N.L., MX':
        return ('MONTERREY')
    else:
        return ('None')
weather_data['Zone'] = weather_data['NAME'].apply(lambda x: label_zone(x))
weather_data.head()
```
